```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv('dataset/GSPC.csv')
```

```python
df = df.drop ('Date',axis = 1)
```

```python
df [-2::]
```

```python
# Armazenando a última linha da base de dados
amanha = df[-1::]
amanha.to_csv ('dataset/futuro.csv', index= False)
```

```python
# Armzenando todas linhas excetos a ultima
base = df.drop( df[-1::].index, axis = 0)
base.tail()
base.to_csv ('dataset/hoje.csv', index= False)
```

```python
# Criando a coluna target com o valor de fechamento do dia seguinte
base['target'] = base['Close'] [1:len(base)].reset_index(drop= True)
base.tail()
```

```python
# Selecionando a ultima linha da bd para a varivel de teste
prev= base[-1::].drop ('target',axis = 1)
prev
```

```python
# selecionando todas linhas excetos as duas ultimas para a base de treino
treino = base.drop (base[-1::].index,axis= 0)
treino.tail()
```

```python
treino.loc [treino ['target'] > treino ['Close'], 'target'] = 1
treino.loc [treino ['target'] != 1, 'target'] = 0
treino.tail()
```

```python
#Separando a base 
y = treino['target']
x = treino.drop ('target', axis = 1)

#Criando os conjuntos de dados de treino e de teste
x_treino,x_teste,y_treino, y_teste = train_test_split (x, y, test_size= 0.3)

#Criação do Modelo
modelo = ExtraTreesClassifier()
modelo.fit (x_treino, y_treino)

# Conferindo a acurácia
resultado = modelo.score (x_teste, y_teste)
print ('Acurácia', resultado)
```

```python
modelo.predict(prev)
```

```python
prev['target'] = modelo.predict (prev)
```

```python
prev.to_csv ('dataset/prev.csv', index= False)
```
